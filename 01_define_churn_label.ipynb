```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# The 'r' before the string tells Python to ignore backslash escape characters
df_gap = pd.read_csv(r"C:\Users\nadim\capstone-churn-analytics\docs\sql_outputs\Supabase Snippet Untitled query (5).csv")

# Plot the distribution of days between first and second purchase
plt.figure(figsize=(10, 6))
sns.histplot(df_gap['gap_days'], bins=50, kde=True)
plt.title('Distribution of Days Between First and Second Order')
plt.xlabel('Days')
plt.ylabel('Number of Customers')
plt.show()

# Calculate percentiles to make your business decision
print("Median gap days:", df_gap['gap_days'].median())
print("75th percentile:", df_gap['gap_days'].quantile(0.75))
print("90th percentile:", df_gap['gap_days'].quantile(0.90))
```
